```python
import json
import pandas as pd
import ast
from matplotlib import pyplot as plt
import os

filename = pd.read_csv('titles_mod.csv')
```

```python
# Creating the box plot of runtime vs genres
dictionary = {}
for index, row in filename.iterrows():
    for genre in set(ast.literal_eval(row['genres'])):
        if genre in dictionary:
            dictionary[genre].append(row['runtime'])
        else:
            dictionary[genre] = [row['runtime']]

genres_count = {}
for genre in dictionary:
    if len(dictionary[genre]) > 0:

        genres_count[genre] = (sum(dictionary[genre]) / len(dictionary[genre]))

with open("runtime_genre.json", "w") as json_file:
    json.dump(genres_count, json_file, indent=4)

#create the bar plot

sort_dict = dict(sorted(genres_count.items(), key=lambda x:x[1], reverse=True))

genres = list(sort_dict.keys())
average_runtime = list(sort_dict.values())

plt.bar(genres, average_runtime)
#plt.title('Average Runtime of All Genres')
plt.xticks(rotation=90)
plt.tight_layout(pad=2)
plt.xlabel('Genre')
plt.ylabel(f'Average Runtime')


plt.savefig('Figures/runtimeVsGenresPlot.png')
```

```python
# creating the box plot of imdb rating vs genres
dictionary = {}
for index, row in filename.iterrows():
    for genre in set(ast.literal_eval(row['genres'])):
        if genre in dictionary:
            dictionary[genre].append(row['imdb_score'])
        else:
            dictionary[genre] = [row['imdb_score']]

genres_count = {}
for genre in dictionary:
    if len(dictionary[genre]) > 0:

        genres_count[genre] = (sum(dictionary[genre]) / len(dictionary[genre]))



with open("imdbRating_genre.json", "w") as json_file:
    json.dump(genres_count, json_file, indent=4)


#create the bar plot

sort_dict = dict(sorted(genres_count.items(), key=lambda x:x[1], reverse=True))

genres = list(sort_dict.keys())
average_rating = list(sort_dict.values())

plt.bar(genres, average_rating)
plt.title('Average Imdb Rating of All Genres')
plt.xticks(rotation=90)
plt.tight_layout(pad=2)
plt.xlabel('Genre')
plt.ylabel(f'Average Imdb Rating')
plt.savefig('Figures/imdbRatingVsGenresPlot.png')
```
